```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# Load the CSV data (when running this notebook, ensure the CSV file path is correct)
df = pd.read_csv('astar_performance_results.csv')

# Display basic information about the dataset
display(Markdown("## A* Pathfinding Performance Data"))
display(Markdown("### Dataset Overview"))
display(df.head())
display(Markdown(f"Total rows: {len(df)}"))

# Calculate average execution times across trials
avg_times = df.groupby(['GridSize', 'Algorithm', 'ThreadCount'])['ExecutionTime'].mean().reset_index()

# Display average execution times
display(Markdown("### Average Execution Times"))
display(avg_times)

# Pivot the data for easier comparison
pivot_df = avg_times.pivot_table(
    index=['GridSize', 'Algorithm'], 
    columns='ThreadCount', 
    values='ExecutionTime'
).reset_index()

display(Markdown("### Pivoted Data (Execution Time by Thread Count)"))
display(pivot_df)

# Calculate speedup relative to sequential version
display(Markdown("## Performance Analysis"))

# Function to calculate speedup
def calculate_speedup(df):
    result_df = pd.DataFrame()
    
    for grid_size in df['GridSize'].unique():
        # Get sequential time for this grid size
        seq_time = df[(df['GridSize'] == grid_size) & 
                      (df['Algorithm'] == 'Sequential') & 
                      (df['ThreadCount'] == 1)]['ExecutionTime'].values[0]
        
        # Calculate speedup for all configurations with this grid size
        grid_df = df[df['GridSize'] == grid_size].copy()
        grid_df['Speedup'] = seq_time / grid_df['ExecutionTime']
        
        result_df = pd.concat([result_df, grid_df])
    
    return result_df

speedup_df = calculate_speedup(avg_times)
display(Markdown("### Speedup Relative to Sequential Version"))
display(speedup_df[speedup_df['Algorithm'] == 'Parallel'])

# Visualization 1: Execution time comparison by grid size
plt.figure(figsize=(15, 10))
for i, size in enumerate(df['GridSize'].unique()):
    plt.subplot(2, 3, i+1)
    
    # Filter data for this grid size
    size_data = avg_times[avg_times['GridSize'] == size]
    
    # Prepare data for plotting
    seq_data = size_data[size_data['Algorithm'] == 'Sequential']
    par_data = size_data[size_data['Algorithm'] == 'Parallel']
    
    # Plot
    plt.bar(0, seq_data['ExecutionTime'].values[0], label='Sequential (1 thread)')
    
    # Get positions for parallel bars
    positions = range(1, len(par_data) + 1)
    thread_counts = par_data['ThreadCount'].values
    exec_times = par_data['ExecutionTime'].values
    
    # Plot parallel bars
    bars = plt.bar(positions, exec_times)
    
    # Add thread count labels
    plt.xticks(
        [0] + list(positions),
        ['Seq (1)'] + [f'Par ({int(tc)})' for tc in thread_counts]
    )
    
    plt.title(f'Grid Size: {size}x{size}')
    plt.ylabel('Execution Time (ms)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
plt.tight_layout()
plt.suptitle('Execution Time Comparison by Grid Size and Thread Count', y=1.02, fontsize=16)
plt.savefig('execution_time_comparison.png')
plt.show()

# Visualization 2: Speedup analysis
plt.figure(figsize=(14, 8))

# Only use parallel data for speedup plot
par_speedup = speedup_df[speedup_df['Algorithm'] == 'Parallel']

# Group by grid size and thread count
for size in df['GridSize'].unique():
    size_data = par_speedup[par_speedup['GridSize'] == size]
    plt.plot(size_data['ThreadCount'], size_data['Speedup'], 
             marker='o', label=f'Grid {size}x{size}')

plt.axhline(y=1, color='r', linestyle='--', label='Sequential baseline')
plt.xlabel('Thread Count')
plt.ylabel('Speedup (Sequential Time / Parallel Time)')
plt.title('Speedup Analysis by Grid Size and Thread Count')
plt.grid(True)
plt.legend()
plt.savefig('speedup_analysis.png')
plt.show()

# Visualization 3: Heatmap of execution times
display(Markdown("### Heatmap of Execution Times"))

# Prepare data for heatmap
heat_df = avg_times.copy()
# Add a combined label for algorithm and thread count
heat_df['Implementation'] = heat_df.apply(
    lambda x: f"{x['Algorithm']} ({x['ThreadCount']} threads)" if x['Algorithm'] == 'Parallel' 
    else 'Sequential (1 thread)', axis=1
)

# Create pivot table
heat_pivot = heat_df.pivot_table(
    index='GridSize', 
    columns='Implementation',
    values='ExecutionTime'
)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heat_pivot, annot=True, fmt='.1f', cmap='YlGnBu')
plt.title('Execution Time (ms) by Grid Size and Implementation')
plt.tight_layout()
plt.savefig('execution_time_heatmap.png')
plt.show()

# Visualization 4: Efficiency analysis
display(Markdown("### Parallel Efficiency Analysis"))

# Calculate efficiency (how well the additional threads are utilized)
efficiency_df = par_speedup.copy()
efficiency_df['Efficiency'] = efficiency_df['Speedup'] / efficiency_df['ThreadCount']

# Plot efficiency
plt.figure(figsize=(14, 8))
for size in df['GridSize'].unique():
    size_data = efficiency_df[efficiency_df['GridSize'] == size]
    plt.plot(size_data['ThreadCount'], size_data['Efficiency'], 
             marker='o', label=f'Grid {size}x{size}')

plt.axhline(y=1, color='r', linestyle='--', label='Perfect efficiency')
plt.xlabel('Thread Count')
plt.ylabel('Efficiency (Speedup / Thread Count)')
plt.title('Parallel Efficiency Analysis')
plt.grid(True)
plt.legend()
plt.savefig('efficiency_analysis.png')
plt.show()

# Statistical analysis
display(Markdown("## Statistical Analysis"))

# Perform t-tests for each grid size to compare sequential vs parallel implementations
from scipy import stats

# Function to perform t-test
def perform_ttest(grid_size, thread_count):
    seq_times = df[(df['GridSize'] == grid_size) & 
                   (df['Algorithm'] == 'Sequential')]['ExecutionTime']
    
    par_times = df[(df['GridSize'] == grid_size) & 
                    (df['Algorithm'] == 'Parallel') & 
                    (df['ThreadCount'] == thread_count)]['ExecutionTime']
    
    t_stat, p_val = stats.ttest_ind(seq_times, par_times)
    
    return {
        'Grid Size': grid_size,
        'Thread Count': thread_count,
        't-statistic': t_stat,
        'p-value': p_val,
        'Significant': p_val < 0.05
    }

# Perform t-tests for all configurations
ttest_results = []
for grid_size in df['GridSize'].unique():
    for thread_count in df[df['Algorithm'] == 'Parallel']['ThreadCount'].unique():
        ttest_results.append(perform_ttest(grid_size, thread_count))

ttest_df = pd.DataFrame(ttest_results)
display(ttest_df)

# Summary and conclusions
display(Markdown("""
## Summary and Conclusions

### Key Findings:

1. **Grid Size Impact**: 
   - For small grid sizes, the overhead of creating and managing threads may outweigh the benefits of parallelization
   - For larger grid sizes, parallel implementation shows significant performance improvements

2. **Thread Count Analysis**:
   - Performance generally improves with more threads, but with diminishing returns
   - The optimal thread count depends on the grid size and available hardware

3. **Efficiency Analysis**:
   - Efficiency decreases as thread count increases, indicating that not all threads are fully utilized
   - Larger grid sizes show better efficiency with higher thread counts

4. **Statistical Significance**:
   - The performance differences between sequential and parallel implementations are statistically significant for larger grid sizes

### Recommendations:

1. For small grid sizes (10x10, possibly 50x50), using the sequential implementation may be more efficient due to lower overhead
2. For medium to large grid sizes, parallel implementation provides better performance
3. The optimal thread count appears to be related to the number of available CPU cores, with diminishing returns beyond that point
4. Further performance optimization could focus on load balancing strategies to improve thread utilization efficiency
"""))
```
